```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
```

```python
# Display grid of 16 circles
def DrawGrid():
    plt.figure(figsize=(6,6))
    [gx, gy] = np.meshgrid([0, 20, 40, 60], [0, 20, 40, 60])
    plt.plot(gx, gy,'o', color='lightgray', markersize=16); plt.axis('square');
    for k,xy in enumerate(zip(gx.flatten(), gy.flatten())):
        os = 0.6 if k+1<10 else 1.1
        plt.text(xy[0]-os, xy[1]-0.7, str(k+1))
```

```python
DrawGrid()
```

## (a) Fit Points with a Spline

```python
# === YOUR CODE HERE ===
t = [0,487,997,1507,2010,2498]
x = [62,46,18,32,41,5]
y = [21,18,34,42,29,15]
spl = make_interp_spline(t, np.c_[x, y])
```

```python
tt = np.linspace(0, 2498, 100000)
xx,yy = spl(tt).T
```

## (b) Plot the Spline

```python
# === YOUR CODE HERE ===
DrawGrid()
plt.plot(x, y, 'o')
plt.plot(xx,yy,'-')
```
